```python
from ortools.linear_solver import pywraplp

def solve_course_assignment(m, n, teacher_preferences, conflicts):
    solver = pywraplp.Solver.CreateSolver('SCIP')
    if not solver:
        return -1

    # Variables
    x = {}
    for t in range(m):
        for c in range(n):
            x[t, c] = solver.BoolVar(f'x[{t},{c}]')
    L = solver.IntVar(0, n, 'L')

    # Constraints
    # Each course is assigned to exactly one teacher
    for c in range(n):
        solver.Add(sum(x[t, c] for t in range(m)) == 1)

    # Teachers can only teach courses in their preference list
    for t in range(m):
        for c in range(n):
            if c + 1 not in teacher_preferences[t]:
                solver.Add(x[t, c] == 0)

    # Conflicting courses cannot be assigned to the same teacher
    for i, j in conflicts:
        for t in range(m):
            solver.Add(x[t, i - 1] + x[t, j - 1] <= 1)

    # Maximal load constraint
    for t in range(m):
        solver.Add(sum(x[t, c] for c in range(n)) <= L)

    # Objective: Minimize the maximal load
    solver.Minimize(L)

    # Solve the problem
    status = solver.Solve()

    if status == pywraplp.Solver.OPTIMAL:
        return int(L.solution_value())
    else:
        return -1

# Input example
m, n = 4, 12
teacher_preferences = [
    [1, 3, 5, 10, 12],
    [9, 3, 4, 8, 12],
    [1, 2, 3, 4, 9, 7],
    [1, 2, 3, 5, 6, 10, 11],
]
conflicts = [
    (1, 2), (1, 3), (1, 5), (2, 4), (2, 5),
    (2, 6), (3, 5), (3, 7), (3, 10), (4, 6),
    (4, 9), (5, 6), (5, 7), (5, 8), (6, 8),
    (6, 9), (7, 8), (7, 10), (7, 11), (8, 9),
    (8, 11), (8, 12), (9, 12), (10, 11), (11, 12)
]

# Solve and print result
result = solve_course_assignment(m, n, teacher_preferences, conflicts)
print(result)
```
